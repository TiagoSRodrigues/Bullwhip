```python
import pymongo

class db_connection():
    def __init__(self):

        self.mongo_client = pymongo.MongoClient("mongodb://localhost:2021/")
        self.simulation_db = self.mongo_client["simulation"]

    def get_collection_data(self, collection_name):
        data=self.simulation_db[collection_name].find()
        for el in data:
            print(el)

    def get_inventories(self):
        main_inventory={}
        filter = {"name": {"$regex": r"^inventory_[0-9].*"}}
        collections_list =self.simulation_db.list_collection_names(filter=filter)

        for col in collections_list:
            cursor= self.simulation_db[col].find()
            data=[]
            for el in cursor:
                data.append(el)
            main_inventory[col]=[data]
        return main_inventory
    
    def get_actor_inventory(self, actor_id):
        main_inventory={}
        filter = {"name": {"$regex": r"^inventory_"+str(actor_id)+".*"}}
        collections_list =self.simulation_db.list_collection_names(filter=filter)

        for col in collections_list:
            cursor= self.simulation_db[col].find()
            data=[]
            for el in cursor:
                data.append(el)
            main_inventory[col]=[data]
        return main_inventory
    
    def get_transactions(self):
            transactions={}
            filter = {"name": {"$regex": r"^closed_transactions"}}
            collections_list =self.simulation_db.list_collection_names(filter=filter)

            for col in collections_list:
                cursor= self.simulation_db[col].find()
                data=[]
                for el in cursor:
                    data.append(el)
                transactions[col]=[data]
            return transactions

    def get_orders(self):
        main_inventory={}
        filter = {"name": {"$regex": r"^orders_[0-9].*"}}
        collections_list =self.simulation_db.list_collection_names(filter=filter)

        for col in collections_list:
            cursor= self.simulation_db[col].find()
            data=[]
            for el in cursor:
                data.append(el)
            main_inventory[col]=[data]
        return main_inventory
    
    def get_actor_orders(self, actor_id):
        orders={}
        filter = {"name": {"$regex": r"^orders_"+str(actor_id)+".*"}}
        collections_list =self.simulation_db.list_collection_names(filter=filter)

        for col in collections_list:
            cursor= self.simulation_db[col].find()
            data=[]
            for el in cursor:
                data.append(el)
            orders[col]=[data]
        return orders
    
    def get_simulation_stats(self):
        stats={}
        filter = {"name": {"$regex": r"^simulation_stats"}}
        collections_list =self.simulation_db.list_collection_names(filter=filter)

        for col in collections_list:
            cursor= self.simulation_db[col].find()
            data=[]
            for el in cursor:
                data.append(el)
            stats[col]=[data]
        return stats       
    
    def get_inventory_snapshot(self):
        
        inventory={}
        filter = {"name": {"$regex": r"^inventory_snapshot"}}
        collections_list =self.simulation_db.list_collection_names(filter=filter)

        for col in collections_list:
            cursor= self.simulation_db[col].find()
            inventory[col]=list(cursor)
                
#         with open ("teste.json","w") as f:
#             f.write(str(inventory).replace("'",'"'))
        return inventory
```
